```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Setting url to automatically pull mushrooms data into a df, using 3 columns for analysis.
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
mushrooms_df = pd.read_csv(url, header = None, usecols = [0, 3, 5], names = ['edibility', 'cap_color', 'odor'])
display(mushrooms_df)
```

```python
# Creating dictionary to hold assigned values for edibility, 0=edible, 1=poisonous.
newvals_dict = {'e' : 0, 'p' : 1}

# This replaces the values for edibility only.
mushrooms_df.replace({'edibility': newvals_dict}, inplace=True)
display(mushrooms_df)
```

```python
# Utilizing same newvals dict for cap_color, assigning number based on first letter of color aka b=Brown.
newvals_dict = {'n' : 0, 'b' : 1, 'c' : 2, 'g' : 3, 'r' : 4, 'p' : 5, 'u' : 6, 'e' : 7, 'w' : 8, 'y' : 9}

# This replaces the values for cap_color only.
mushrooms_df.replace({'cap_color': newvals_dict}, inplace=True)
display(mushrooms_df)
```

```python
# Utilizing newvals dict for odor, assigning number based on first letter of odor when applicable aka a=Almond.
newvals_dict = {'a' : 0, 'l' : 1, 'c' : 2, 'y' : 3, 'f' : 4, 'm' : 5, 'n' : 6, 'p' : 7, 's' : 8}

# This replaces the values for odor only.
mushrooms_df.replace({'odor': newvals_dict}, inplace=True)
display(mushrooms_df)
```

```python
# Creating a scatterplot to show odor/edibility relationship, utilizing dark settings.
sns.set(style='darkgrid')
sns.set_color_codes('dark')
sns.lmplot('odor', 'edibility', data=mushrooms_df)
```

```python
# Creating a scatterplot to show cap_color/edibility relationship, utilizing dark settings as above.
sns.lmplot('cap_color', 'edibility', data=mushrooms_df)
```

# Conclusion
### Our results from the above scatterplot analysis shows that the odor of the mushroom could be used as an indicator if the mushroom is edible or not (Anise/Almond = edible, while all other odors have a degree of plausibility to be poisonous). The following scatterplot analysis of the mushroom's cap_color can also be used to determine if the mushroom is edible or not due to Purple and Green cap_color mushrooms are not poisonous, but all others should be avoided due to their degree of plausibility of being poisonous.
